## 10.Visualisation

### Scatter Plots

```python
sns.pairplot(df,hue='class')
plt.show()
```

### Visualizing Distributions(Histplot)

```python
plt.figure(figsize=(12, 6))
for i, col in enumerate(df.columns[:-1]):
    plt.subplot(2, 2, i + 1)
    sns.histplot(df[col], kde=True)
    plt.title(f'{col} Distribution')
plt.tight_layout()
plt.show()
```

### Box Plots(Identify outliers)

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, orient='h')
plt.title('Box Plots for Iris Features')
plt.show()
```

### Violin Plots

```python

fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Violin Plots of Features by Class', fontsize=16)

for i, col in enumerate(df.columns[:-1]):
    row = i // 2
    col_pos = i % 2
    sns.violinplot(  data=df, x='class', y=col,ax=axes[row, col_pos])
    axes[row, col_pos].set_title(f'Violin Plot of {col} by Class')

plt.tight_layout()  # Adjust layout to fit the title
plt.show()
```

### Heatmap(Correlations)

```python
plt.figure(figsize=(10, 8))

# Exclude non-numeric columns (species)
numeric_df = df.select_dtypes(include='number')

# Calculating the correlation matrix
corr_matrix = numeric_df.corr()

# Visualizing the Correlation Matrix
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

### Outlier Treatment

```python
#box plot to identify outliers
fig = plt.figure(figsize=(20,20))
counter = 0

for col in df.select_dtypes(include="number").columns:
    fig.add_subplot(4,4,counter+1)
    sns.boxplot(data=df[col])
    plt.xlabel(col)
    counter = counter + 1
plt.show()
```

```python
def whisker(col):
    q1,q3=np.percentile(col,[25,75])
    iqr=q3-q1
    lw=q1-1.5*iqr
    uw=q3+1.5*iqr
    return lw,uw
```

```python
#removing outliers using capping
for i in df.select_dtypes(include="number").columns:
    lw,uw=whisker(df[i])
    df[i]=np.where(df[i]<lw,lw,df[i])
    df[i]=np.where(df[i]>uw,uw,df[i]) 

```

```python
#box plot after outlier treatment
fig = plt.figure(figsize=(20,20))
counter = 0

for col in df.select_dtypes(include="number").columns:
    fig.add_subplot(4,4,counter+1)
    sns.boxplot(data=df[col])
    plt.xlabel(col)
    counter = counter + 1
plt.show()    
    
```

## 11.Feature Engineering

### Label Encoding

```python
from sklearn.preprocessing import LabelEncoder

# Assuming 'class' is the categorical column in your dataset
label_encoder = LabelEncoder()
df['class'] = label_encoder.fit_transform(df['class'])  # Directly encoding the column

print("Encoded Classes:", list(label_encoder.classes_))
print(df.head())  # Displaying the encoded DataFrame
```

```python
df
```

### Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler

# Select numerical columns for scaling
columns=[ 'sepal length','sepal width',	'petal length',	'petal width']

# Create a scaler object
scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(df[columns])

# Convert scaled data to DataFrame
scaled_df = pd.DataFrame(df_scaled, columns=columns)

df[columns] = scaled_df

df
```
